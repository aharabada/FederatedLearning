```python
import os
import numpy as np
import pickle

from DataCollector import CollectedData

# since this Notebook is in the code folder, we need to go up one level, so every module works as expected
if(os.getcwd().endswith("code")):
    os.chdir("..")
print(os.getcwd())
```

```python
for root, dirnames, filenames in os.walk("results"):
    for file in filenames:
        if file.endswith(".pkl") and file.startswith("FedSTD"):
            print(f"Loading {file}")
            data = pickle.load(open(f"{root}/{file}", "rb"))
            for key, value in data.plot_data.items():
                mean = np.mean(value, axis=0)
                std = np.std(value, axis=0)
                print("=" * 50)
                print(f"{file.removesuffix('_collected_data.pkl')} - {key}")
                print("=" * 50)
                for i in range(len(mean)):
                    print(f"Step {i}: {mean[i]:.4f} \\\\ $\pm$ {std[i]:.4f}")
                
```
